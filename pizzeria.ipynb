***Inventory Management***

Total quantity of each item.

```python
%%sql
SELECT
    o.item_id,
    i.sku,
    i.item_size,
    i.item_name,
    sum(o.quantity) as order_quantity
FROM 
    orders o
    LEFT JOIN item i ON o.item_id = i.item_id
GROUP BY o.item_id, i.sku, i.item_name
```

Total quantity by ingredient.

```python
%%sql
SELECT
    o.item_id,
    i.sku,
    i.item_name,
    r.ing_id,
    ing.ing_name,
    r.quantity AS recipe_quantity,
    sum(o.quantity) as order_quantity
FROM 
    orders o
    LEFT JOIN item i ON o.item_id = i.item_id
    LEFT JOIN recipe r ON i.sku = r.recipe_id
    LEFT JOIN ingredients ing ON ing.ing_id = r.ing_id
GROUP BY 
    o.item_id, 
    i.sku, 
    i.item_name,
    r.ing_id,
    r.quantity,
    ing.ing_name
LIMIT 10
```

Total cost of ingredients.

```python
%%sql
SELECT
    o.item_id,
    i.sku,
    i.item_name,
    r.ing_id,
    ing.ing_name,
    r.quantity AS recipe_quantity,
    sum(o.quantity) as order_quantity,
    ing.ing_weight,
    ing.ing_price
FROM 
    orders o
    LEFT JOIN item i ON o.item_id = i.item_id
    LEFT JOIN recipe r ON i.sku = r.recipe_id
    LEFT JOIN ingredients ing ON ing.ing_id = r.ing_id
GROUP BY 
    o.item_id, 
    i.sku, 
    i.item_name,
    r.ing_id,
    r.quantity,
    ing.ing_name,
    ing.ing_weight,
    ing.ing_price
LIMIT 10
```

Total cost of each pizza.

Now want to multiply recipe_quantity by order_quantity but order_quantity is the result of a sum function. Therefore, need a new select statement; have to use a sub-query. 

Given whole query above an alias of s1. Able to select only some columns while all columns from s1 remain available if want to view in future. 

```python
%%sql
SELECT
    s1.sku,
    s1.ing_id,
    s1.ing_name,
    s1.order_quantity,
    s1.recipe_quantity,
    s1.order_quantity*s1.recipe_quantity AS ordered_weight,
    s1.ing_price/s1.ing_weight AS unit_cost,
    (s1.order_quantity*s1.recipe_quantity)*(s1.ing_price/s1.ing_weight) AS ingredient_cost
FROM (SELECT
    o.item_id,
    i.sku,
    i.item_name,
    r.ing_id,
    ing.ing_name,
    r.quantity AS recipe_quantity,
    sum(o.quantity) as order_quantity,
    ing.ing_weight,
    ing.ing_price
FROM 
    orders o
    LEFT JOIN item i ON o.item_id = i.item_id
    LEFT JOIN recipe r ON i.sku = r.recipe_id
    LEFT JOIN ingredients ing ON ing.ing_id = r.ing_id
GROUP BY 
    o.item_id, 
    i.sku, 
    i.item_name,
    r.ing_id,
    r.quantity,
    ing.ing_name,
    ing.ing_weight,
    ing.ing_price) s1
LIMIT 20
```

Total weight used of each ingredient.

Save everything have so far as a view called stock_1. Have to make sure output not limited to a set number of rows, otherwise view will only be that set length.

```python
%%sql
CREATE VIEW stock_1 AS
SELECT
    s1.sku,
    s1.ing_id,
    s1.ing_name,
    s1.order_quantity,
    s1.recipe_quantity,
    s1.order_quantity*s1.recipe_quantity AS ordered_weight,
    s1.ing_price/s1.ing_weight AS unit_cost,
    (s1.order_quantity*s1.recipe_quantity)*(s1.ing_price/s1.ing_weight) AS ingredient_cost
FROM (SELECT
    o.item_id,
    i.sku,
    i.item_name,
    r.ing_id,
    ing.ing_name,
    r.quantity AS recipe_quantity,
    sum(o.quantity) as order_quantity,
    ing.ing_weight,
    ing.ing_price
FROM 
    orders o
    LEFT JOIN item i ON o.item_id = i.item_id
    LEFT JOIN recipe r ON i.sku = r.recipe_id
    LEFT JOIN ingredients ing ON ing.ing_id = r.ing_id
GROUP BY 
    o.item_id, 
    i.sku, 
    i.item_name,
    r.ing_id,
    r.quantity,
    ing.ing_name,
    ing.ing_weight,
    ing.ing_price) s1
```

```python
%%sql
SELECT 
    ing_name,
    sum(ordered_weight) AS ordered_weight
FROM 
    stock_1
GROUP BY ing_name
```

Total weight of inventory of each ingredient and how much is remaining.

```python
%%sql
SELECT 
    s2.ing_name,
    s2.ordered_weight,
    ing.ing_weight*inv.quantity as total_inv_weight,
    (ing.ing_weight*inv.quantity)-s2.ordered_weight as remaining_weight
FROM (SELECT 
    ing_id,
    ing_name,
    sum(ordered_weight) AS ordered_weight
FROM 
    stock_1
GROUP BY ing_name, ing_id) s2
LEFT JOIN inventory inv ON inv.item_id = s2.ing_id
LEFT JOIN ingredients ing ON ing.ing_id = s2.ing_id
```

***Staff Management***

View staff schedule.

```python
%%sql
SELECT
    ro.date,
    s.first_name,
    s.last_name,
    s.hourly_rate,
    sh.start_time,
    sh.end_time
FROM rota ro
LEFT JOIN staff s ON ro.staff_id = s.staff_id
LEFT JOIN shift sh ON ro.shift_id = sh.shift_id
LIMIT 20
```

Calculate cost of staff.

```python
%%sql
SELECT
    ro.date,
    s.first_name,
    s.last_name,
    s.hourly_rate,
    sh.start_time,
    sh.end_time,
    CAST((strftime('%s', end_time) - strftime('%s', start_time)) AS REAL) / 3600 AS hours_in_shift,
    CAST((strftime('%s', end_time) - strftime('%s', start_time)) AS REAL) / 3600 * s.hourly_rate AS staff_cost
FROM rota ro
LEFT JOIN staff s ON ro.staff_id = s.staff_id
LEFT JOIN shift sh ON ro.shift_id = sh.shift_id
LIMIT 30
```
